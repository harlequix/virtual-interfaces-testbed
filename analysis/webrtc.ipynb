```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_webrtc_stats_json(file):
    with open(file) as f:
        for line in f:
            yield json.loads(line)
```

```python
stats_types = [
    "codec",
    "inbound-rtp",
    "outbound-rtp",
    "remote-inbound-rtp",
    "remote-outbound-rtp",
    "media-source",
    "media-playout",
    "peer-connection",
    "data-channel",
    "transport",
    "candidate-pair",
    "local-candidate",
    "remote-candidate",
    "certificate"
]
```

```python
c0stats = [x for x in read_webrtc_stats_json('../webrtc/media/get_stats_client_0.json')]
c0stats_by_type = {t: [x for xs in c0stats for x in xs if x['type'] == t] for t in stats_types}

c1stats = [x for x in read_webrtc_stats_json('../webrtc/media/get_stats_client_1.json')]
c1stats_by_type = {t: [x for xs in c1stats for x in xs if x['type'] == t] for t in stats_types}
```

```python
c0_outbound_rtp = pd.DataFrame(c0stats_by_type['outbound-rtp'])
c0_outbound_rtp['targetBitrate'].plot()
```
